## Same sample cross region DEG

```python
st_cells = [x for x in adata.obs['annot'].unique() if 'ST' in x]
ref = 'cortex'
mks = {}
for reg in adata.obs['batch'].unique():
    ad = adata[(adata.obs['batch']==reg)&(adata.obs['annot'].isin(st_cells))]
    print(ad.obs.batch.unique(), ad.shape)
    sc.tl.rank_genes_groups(ad, groupby='region', reference=ref, pts=True, use_raw=False)
    mk = sc.get.rank_genes_groups_df(ad, group=None)
    mks[reg] = mk
sc.tl.rank_genes_groups(adata, groupby='region', reference=ref, pts=True, use_raw=False)
mk = sc.get.rank_genes_groups_df(adata, group=None)
mks['all'] = mk
```

```python
top_mks = {k:v[(v['logfoldchanges']>1)&(v['pvals_adj']<0.05)].sort_values('logfoldchanges', ascending=False) for k,v in mks.items()}
batch_separated_mks = {}
for k,v in top_mks.items():
    print('Region', k)
    regs = v['group'].unique()
    batch_inRegion_mks = {reg: v[v['group']==reg].sort_values('pvals_adj', ascending=True) for reg in regs}
    batch_separated_mks[k] = batch_inRegion_mks
flattened = {}
for region in batch_separated_mks.keys():
    for sample in batch_separated_mks[region].keys():
        print('REGION', region, 'SAMPLE', sample)
        cortex_slow = batch_separated_mks[region][sample]
        print('Ribo genes in top 30 degs', [x for x in cortex_slow.head(100)['names'] if 'RPS' in x or 'RPL' in x])
        rg = [x for x in cortex_slow['names'] if 'RPS' in x or 'RPL' in x]
        flattened['_'.join([region, sample])] = rg
        perc = len(rg)/len(cortex_slow['names'].unique())
        print('Ribo genes as percent of total deg compared to other {region}:', perc)
```

```python
for k,v in batch_separated_mks.items():
    top_dict = {i:df['names'][:10] for i,df in v.items()}
    ad = adata[adata.obs['batch']==k]
    sc.pl.dotplot(ad, var_names=top_dict, groupby='region', use_raw=False)
```

```python
genes = list(set(flattened['all_mid-cortex']).intersection(set(flattened['all_cortex'])))
print(len(genes)/len(set(flattened['all_mid-cortex'])), len(genes)/len(set(flattened['all_cortex'])), )
# 0.7608695652173914 0.5982905982905983
sc.pl.dotplot(adata, var_names=genes, groupby='region', use_raw=False)
```

## Ribo score based on all ribo genes

```python
genes = [x for x in adata.var_names if x[:3] in ['RPS', 'RPL'] and '-' not in x and all([True if t not in x else False for t in ['BP']])]
genes
```

```python
sc.tl.score_genes(adata, genes, use_raw=False, score_name='ribo')

from scipy.stats import ranksums
import seaborn as sns
from statannotations.Annotator import Annotator

data = adata.obs[['region', 'batch', 'ribo']]
data['batch-region'] = [f'{x}_{data.region[i]}' for i,x in enumerate(data['batch'])]
```

```python
x = 'region'
y = 'ribo'
order = ['medulla', 'mid-cortex', 'cortex']
# custom_test = StatTest(ranksums(alternative=['greater']))
ax = sns.boxplot(data=data, x=x, y=y, order=order)
pairs=[("mid-cortex", "medulla"), ("cortex", "medulla"), ("cortex", "mid-cortex")]
annotator = Annotator(ax, pairs, data=data, x=x, y=y, order=order)
annotator.configure(test='Mann-Whitney', text_format='star', 
                    loc='outside')
annotator.apply_test(alternative='greater')
annotator.apply_and_annotate()
plt.show()

x = 'batch'
y = 'ribo'
order = ['Fresh', 'Slow', 'Vitri']
# custom_test = StatTest(ranksums(alternative=['greater']))
ax = sns.boxplot(data=data, x=x, y=y, order=order)
pairs=[( "Fresh", "Slow"), ("Fresh", "Vitri"), ('Slow', 'Vitri')]
t1, pval1 = ranksums(data[data['batch']=='Fresh']['ribo'], data[data['batch']=='Slow']['ribo'], alternative='greater')
print('test Fresh vs Slow', t1, pval1)
t2, pval2 = ranksums(data[data['batch']=='Fresh']['ribo'], data[data['batch']=='Vitri']['ribo'], alternative='greater')
print('test Fresh vs Vitri', t2, pval2)
t3, pval3 = ranksums(data[data['batch']=='Slow']['ribo'], data[data['batch']=='Vitri']['ribo'], alternative='greater')
annotator = Annotator(ax, pairs, data=data, x=x, y=y, order=order)
# custom_test = StatTest(ranksums(alternative='greater'), test_long_name='Wilcoxon', test_short_name='Wilc.')
annotator.configure(test='Mann-Whitney', text_format='star', 
                    loc='outside')
annotator.apply_and_annotate()
plt.text(0.8, 0.5, f'Alt: greater\nFresh vs Slow pval={pval1}\nFresh vs Vitri pval={pval2}\nSlow vs Vitri pval={pval3}', fontsize=10, ha='center')
plt.show()
```

```python
figdir = 'fig'
x = 'batch'
y = 'ribo'
for reg in data.region.unique():
    dcpy = data.copy()
    data = data[data['region'].isin([reg])]
    order = ['Fresh', 'Slow', 'Vitri']
    ax = sns.boxplot(data=data, x=x, y=y, order=order)
    pairs=[( "Fresh", "Slow"), ("Fresh", "Vitri"), ('Slow', 'Vitri')]
    t1, pval1 = ranksums(data[data['batch']=='Fresh']['ribo'], data[data['batch']=='Slow']['ribo'], alternative='less')
    print('test Fresh vs Slow', t1, pval1)
    t2, pval2 = ranksums(data[data['batch']=='Fresh']['ribo'], data[data['batch']=='Vitri']['ribo'], alternative='less')
    print('test Fresh vs Vitri', t2, pval2)
    t3, pval3 = ranksums(data[data['batch']=='Slow']['ribo'], data[data['batch']=='Vitri']['ribo'], alternative='less')
    print('test Slow vs Vitri', t3, pval3)
    annotator = Annotator(ax, pairs, data=data, x=x, y=y, order=order)
    # custom_test = StatTest(ranksums(alternative='greater'), test_long_name='Wilcoxon', test_short_name='Wilc.')
    annotator.configure(test='Mann-Whitney', text_format='star', 
                        loc='inside')
    annotator.apply_and_annotate()
    plt.text(0.8, 0.2, f'Region: {reg}\nAlt: less\nFresh vs Slow pval={pval1}\nFresh vs Vitri pval={pval2}\nSlow vs Vitri pval={pval3}', fontsize=10, ha='center')
    plt.savefig(f'{figdir}/ribo_{reg}_cross_{x}.svg')
    plt.show()
    data = dcpy
```

## Same region cross sample DEG

```python
st_cells = [x for x in adata.obs['annot'].unique() if 'ST' in x]
print(st_cells)
ref = 'rest'
mks = {}
for reg in adata.obs['region'].unique():
    ad = adata[(adata.obs['region']==reg)&(adata.obs['annot'].isin(st_cells))]
    res = []
    for b in ad.obs['batch'].unique():
        print(b)
        ref = 'Fresh' if b in ['Vitri', 'Slow'] else 'rest'
        sc.tl.rank_genes_groups(ad, groupby='batch', reference=ref, pts=True, use_raw=False)
        mk = sc.get.rank_genes_groups_df(ad, group=None)
        mk = mk[mk['group']==b]
        res.append(mk)
    mk = pd.concat(res)
    mks[reg] = mk
```

```python
for k,v in mks.items():
    df = v[(v['pvals_adj']<0.05)&(v['pct_nz_group']>0.2)].sort_values('scores', ascending=False)
    df = df[df['logfoldchanges']>1]
    df = df[~df['names'].str.contains('MT-|LINC|AC|FP|KA|KB|KC|BP')]   
    top_dict = {grp:df[df['group']==grp]['names'][:10] for grp in df['group'].unique()}
    ad = adata[adata.obs['region']==k]
    sc.pl.dotplot(ad, var_names=top_dict, groupby='batch', use_raw=False, title=k, save=f'{k}_cross_sample_deg.svg')
```

```python
top_mks = {k:v[(v['logfoldchanges']>1)&(v['pvals_adj']<0.05)].sort_values('logfoldchanges', ascending=False) for k,v in mks.items()}
batch_separated_mks = {}
for k,v in top_mks.items():
    print('Region', k)
    regs = v['group'].unique()
    batch_inRegion_mks = {reg: v[v['group']==reg].sort_values('pvals_adj', ascending=True) for reg in regs}
    batch_separated_mks[k] = batch_inRegion_mks
flattened = {}
for region in batch_separated_mks.keys():
    for sample in batch_separated_mks[region].keys():
        print('REGION', region, 'SAMPLE', sample)
        cortex_slow = batch_separated_mks[region][sample]
        rg = [x for x in cortex_slow['names'] if x[:3] in ['RPS', 'RPL'] and '-' not in x and all([True if t not in x else False for t in ['BP','KA', 'KB', 'KC']])]
        print(rg)
        flattened['_'.join([region, sample])] = rg
        perc = len(rg)/len(cortex_slow['names'].unique())
        print('Ribo genes as percent of total deg compared to other {region}:', perc)
        
```

```python
sets = {}
for b in ['Slow', 'Fresh', 'Vitri']:
    slow_ribos = np.unique([x for k,v in flattened.items() for x in v if b in k])
    slow_ribo_counts = {k:len(v) for k,v in flattened.items() if b in k}
    print(b, slow_ribos, slow_ribo_counts)
    sets[b] = set(slow_ribos)
    
```
